```python
# Directories
# Path to raw data
file_in   = os.path.normpath(r'C:\Users\new356\Downloads\SBB\SBB\sasp_eval.nc')
```

```python
# Set font size
font = {'weight' : 'bold',
        'size'   : 20}
matplotlib.rc('font', **font)
```

```python
# Import stations of interest
CRHO_data = xr.open_dataset(file_in,engine='netcdf4')
```

```python
CRHO_data
```

```python
# Sort variables by alpha
AllVars = sorted(CRHO_data.data_vars)
```

```python
# Remove vars not to plot'
[AllVars.remove(cV) for cV in ['year','month','day','hour']]
```

```python
for cvar in AllVars:
    print(cvar)
    plt.figure()
    CRHO_data[cvar].plot()
    plt.title(cvar)
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5,forward=True)
    plt.ylabel(CRHO_data[cvar].attrs['units'])
    plt.axis("tight");
```
